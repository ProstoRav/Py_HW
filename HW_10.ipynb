````python
import pandas as pd
````

````python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
````

````python
pd.get_dummies(data['whoAmI']).head()
````

````python
data = pd.DataFrame({'human': data['whoAmI'] == 'human', 'robot': data['whoAmI'] == 'robot'}).astype(int)
data.head()
````
